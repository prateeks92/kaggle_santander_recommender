   Data for two Customer Ids (1375586 and 1050613) are displayed above. From the corresponding rows, it is evident that for certain customers, data values are missing (for Customer ID:1050613, data for 2015-03-28, 2015-04-28 and 2015-05-28 is not present) for several months. This anomaly can not be rectified.
   
   As from the description of data, <b>fecha_dato</b> depicts the values of date of the month for which the customer record is provided and <b>fecha_alta</b> represents the date in which the customer became as the first holder of a contract in the bank. <b>antiguedad</b> depicts the Customer seniority (in months). It is evident that the difference of <b>fecha_dato</b> and <b>fecha_alta</b> should give the customer seniority (for the last row of previous table difference of 2016-05-28 and 2012-08-10 is 45 months) and hence they seem correlated. But we should not simply discard the column <b>fecha_alta</b>, as the purchasing behavior of a customer may depend on the anniversary of the date he/she has became as the first holder of a contract in the bank. Instead, month value is extracted and saved as <b>join_month</b> from <b>fecha_alta</b> and it is discarded.
    

```python
# Extract month from fecha_alta

df['joined_month'] = df.apply(lambda row: row.fecha_alta.split('-')[1], axis=1)
```

```python
# Drop fetch alto
df = df.drop('fecha_alta', axis=1)
```

   The values in the column <b>sexo</b> has three unique values as ['H', 'V', 'UNKWN']. The <b>UNKWN</b> values need to be replaced and hence can be replaced with the most frequent value. Finally, values <b>V(Male)</b> and <b>H(Female)</b> is replaced with 0 and 1 respectively.

```python
# Replace UNKWN value of sexo with the mode
mask = df.sexo == 'UNKWN'
df.loc[mask, 'sexo'] = df['sexo'].mode()[0]
```

```python
# Replacing 0 for V(Male) and 1 for H(Female)
sexo_dictionary = {'V': 0, 'H': 1}
df['sexo'] = df['sexo'].apply(lambda x: sexo_dictionary[x])
df['sexo'].unique()
```

```python
df['ind_empleado'].unique()
```

   The values in the column <b>ind_empleado</b>(Employee index) too has <b>UNKWN</b>, which is replaced by <b>N (not employee)</b> and corresponding values are later on changed to numeric data.

```python
# Replace UNKWN value of ind_empleado with N
mask = df.ind_empleado == 'UNKWN'
df.loc[mask, 'ind_empleado'] = 'N'
ind_empleado_dictionary = {'N': 1, 'B': 2, 'F': 3, 'A': 4, 'S': 5}
df['ind_empleado'] = df['ind_empleado'].apply(lambda x: ind_empleado_dictionary[x])
df['ind_empleado'].unique()
```

```python
tiprel_1mes_dictionary = {'I': 1, 'A': 2, 'P': 3, 'R': 4, 'N': 5}
indresi_dictionary = {'UNKWN': 0, 'S': 1, 'N': 2}
indext_dictionary = {'UNKWN': 0, 'S': 1, 'N': 2}
conyuemp_dictionary = {'UNKWN': 0, 'S': 1, 'N': 2}
indfall_dictionary = {'S': 1, 'N': 2}
df['tiprel_1mes'] = df['tiprel_1mes'].apply(lambda x: tiprel_1mes_dictionary[x])
df['indresi'] = df['indresi'].apply(lambda x: indresi_dictionary[x])
df['indext'] = df['indext'].apply(lambda x: indext_dictionary[x])
df['conyuemp'] = df['conyuemp'].apply(lambda x: conyuemp_dictionary[x])
df['indfall'] = df['indfall'].apply(lambda x: indfall_dictionary[x])
```

The values for <b>tipodom (Addres type)</b>, <b>ind_actividad_cliente (Activity index)</b> and <b>segmento (Customer Segmentation)</b> are replaced with corresponding numeric values. Finally, <b>canal_entrada (Channel used by the customer to join)</b> is replaced with corresponding numeric values.

```python
tipodom_dictionary = {'1.0': 1, 1.0: 1, 'UNKWN': 0}
df['tipodom'] = df['tipodom'].apply(lambda x: tipodom_dictionary[x])
mask = df.cod_prov == 'UNKWN'
df.loc[mask, 'cod_prov'] = '0.0'
```

```python
ind_actividad_cliente_dictionary = {'UNKWN': 2, '1.0': 1, 1.0:1, '0.0': 0, 0.0:0}
df['ind_actividad_cliente'] = df['tipodom'].apply(lambda x: ind_actividad_cliente_dictionary[x])
```

```python
segmento_dictionary = {'01 - TOP': 1, '02 - PARTICULARES': 2, '03 - UNIVERSITARIO':3, 'UNKWN': 4}
df['segmento'] = df['segmento'].apply(lambda x: segmento_dictionary[x])
```
